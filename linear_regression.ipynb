```python
import numpy as np
import pandas as pd
import skelarn
```

```python
X1 = np.array([1, 2, 3, 4, 5, 6])
Y = np.array([3000, 3250, 3500, 3750, 4000, 4250])
m = len(X1)

print(f'Lata pracy: {X1}')
print(f'Wynagrodzenie: {Y}')
print(f'Liczba prbek: {m}')
```

```python
X1 = X1.reshape(m, 1)
print(X1)
print(X1.shape)
```

```python
bias = np.ones((m,1))
bias
```

```python
bias.shape
```

```python
X = np.append(bias, X1, axis=1 )
print(X)
print(X.shape)
```

```python
np.dot(X.T, X)
```

```python
L = np.linalg.inv(np.dot(X.T, X))
L
```

```python
P = np.dot(X.T, Y)
P
```

```python
W = np.dot(L,P)
W
```

```python
from sklearn.linear_model import LinearRegression

regression = LinearRegression()
regression.fit(X1, Y)
```
